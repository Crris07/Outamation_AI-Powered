```python
#Extract Loan ID parts using groups
match = re.search(r"(L-\d{4})-(\d{4})", text)

if match:
    year = match.group(1)  # 'L-2025'
    sequence = match.group(2)  # '0042'
    print(f"Loan Year: {year}, Sequence: {sequence}")
```
