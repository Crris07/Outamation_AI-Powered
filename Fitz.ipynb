```python
file_1 = '/content/drive/MyDrive/sample_resume.pdf'
```

```python
import fitz
doc = fitz.open(file_1)
print(len(doc))
doc.metadata
```

```python
page = doc[0]
print(page.get_text('text'))
```

```python
words = page.get_text("words")

for w in words[:5]:
  print(w)
```

```python
blocks = page.get_text("blocks")
for b in blocks:
  print(b)
```

```python
for page_num in range(len(doc)):
    page = doc[page_num]
    text = page.get_text("text")
    print(f"--- Page {page_num + 1} ---")
    print(text)
```
